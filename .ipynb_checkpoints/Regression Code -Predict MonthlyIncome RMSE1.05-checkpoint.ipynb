```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
from numpy.linalg import inv
from sklearn.model_selection import train_test_split
from scipy.stats import multivariate_normal
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.utils.multiclass import unique_labels
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r".\dataset\raw\HR_Process_Dataset.csv")
```

```python
df
```

```python
X=df.drop(columns=['EmployeeID','MonthlyIncome'],axis=1)
X
```

```python
y=df['MonthlyIncome']
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y,test_size=0.2,random_state=43)
```

```python
numerical_cols = ['HourlnWeek', 'SalaryHikelastYear', 'Age', 'HomeToWork']
Xtrain_categorical = Xtrain.drop(columns=numerical_cols)
Xtest_categorical = Xtest.drop(columns=numerical_cols)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain[numerical_cols])
Xtest_scaled = scaler.transform(Xtest[numerical_cols])
```

```python
# Convert scaled NumPy arrays back to DataFrames with correct column names and index
Xtrain_scaled_df = pd.DataFrame(Xtrain_scaled, columns=numerical_cols, index=Xtrain.index)
Xtest_scaled_df = pd.DataFrame(Xtest_scaled, columns=numerical_cols, index=Xtest.index)
```

```python
Xtrain_final = pd.concat([Xtrain_scaled_df, Xtrain_categorical], axis=1)
Xtest_final = pd.concat([Xtest_scaled_df, Xtest_categorical], axis=1)
```

```python
scaler = MinMaxScaler()
ytrain_final = scaler.fit_transform(ytrain.values.reshape(-1, 1))
ytest_final = scaler.transform(ytest.values.reshape(-1, 1))
```

```python
# Convert scaled NumPy arrays back to DataFrames with correct column names and index
ytrain_scaled_df = pd.DataFrame(ytrain_final, index=ytrain.index, columns=['MonthlyIncome'])
ytest_scaled_df = pd.DataFrame(ytest_final, index=ytest.index, columns=['MonthlyIncome'])
```

```python
ytrain_scaled_df
```

```python
Xtrain_final = Xtrain_final.sort_index()
ytrain_scaled_df = ytrain_scaled_df.sort_index()
```

```python
Xtrain_scaled_m = np.hstack((np.ones_like(Xtrain_final.values[:,:1]), Xtrain_final.values))
Xtest_scaled_m = np.hstack((np.ones_like(Xtest_final.values[:,:1]), Xtest_final.values))
Xtrain_scaled_m.shape
```

```python
np.log(ytrain_scaled_df.values)
```

```python
w = np.linalg.solve(np.dot(Xtrain_scaled_m.T, Xtrain_scaled_m), np.dot(Xtrain_scaled_m.T, np.log1p(ytrain_scaled_df.values)))
print(w)
```

```python
w=w.reshape(32,)
```

```python
def predict(x, w):
    X_p = np.hstack((np.ones_like(x[:,:1]), x))
    return np.exp(np.dot(X_p ,w))
```

```python
def plot_fit(x_test, y_test, x, y):
    plt.plot(x_test[:,:1], y_test, 'b:')
    plt.plot(x[:,:1], y, 'g:')
    plt.ylabel("y (Age)")
    plt.xlabel("x (Feature)")
    plt.show()


plot_fit(Xtest_final.values, predict(Xtest_final.values, w), Xtrain_final.values, ytrain_scaled_df.values)
```

```python
def compute_RSS(x, y, w):
    return np.sqrt(np.mean((y - predict(x, w)) ** 2))
    

print(compute_RSS(Xtest_final.values, ytest_scaled_df.values,w))
```
