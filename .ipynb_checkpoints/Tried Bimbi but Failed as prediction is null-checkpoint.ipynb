```python
import arviz as az
import bambi as bmb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import xarray as xr
```

```python
df = pd.read_csv(r".\dataset\raw\CleanDataset.csv")
```

```python
df
```

```python
model=bmb.Model("Attrition~Age+HourlnWeek+WorkLifeBalance",data=df)
fitted = model.fit(tune=100, draws=100,chains=2, target_accept=0.95, random_seed=34)
```

```python
model.plot_priors()
```

```python
az.plot_trace(fitted)
```

```python
df1=df[['Age','HourlnWeek','WorkLifeBalance']]
```

```python
df1
```

```python
pred=model.predict(fitted,data=df1,kind='pps')
```

```python
print(pred)
```
